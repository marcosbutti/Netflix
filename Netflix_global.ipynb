#### Datos de Visualizaciones de Netflix

```python
import pandas as pd
import numpy as np

data = pd.read_excel("all-weeks-global.xlsx")
data
```

```python
data.info()
```

```python
horasem = data.sort_values(by='weekly_hours_viewed', ascending=False)
horasem.head(15)
```

```python
vistasem = data.sort_values(by='weekly_views', ascending=False)
vistasem.head(15)
```

```python
vistasem = vistasem.drop_duplicates(subset='show_title', keep='first')
vistasem.head(10)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Calcular los pesos de las palabras basados en los valores numÃ©ricos
weights = dict(zip(vistasem['show_title'], vistasem['weekly_views']))
# Crear la nube de palabras
wc = WordCloud(width = 1000, height = 300, background_color='black', colormap = "Reds").generate_from_frequencies(weights)

# Mostrar la nube de palabras
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')

# Guardar la nube de palabras
#wc.to_file("VistasSemanales.png")
```
