### Datos de visualizaciones de Netflix

```python
import pandas as pd
import numpy as np

data = pd.read_excel("all-weeks-countries.xlsx")
data
```

```python
data.info()
```

```python
data.sort_values(by='cumulative_weeks_in_top_10', ascending=False)
```

```python
data[(data.cumulative_weeks_in_top_10 > 60) & (data.country_iso2 == "AR")]
```

```python
filtrado = data[(data.cumulative_weeks_in_top_10 > 4) & (data.country_iso2 == "AR") & (data.category == "Films")]
filtrado
```

```python
# Filtro para incluir solo los registros con valores únicos
norepetidos = filtrado[(filtrado['show_title'].notnull()) & (~filtrado['show_title'].astype(str).duplicated())]
norepetidos = norepetidos.sort_values(by='cumulative_weeks_in_top_10', ascending=False)
norepetidos
```

```python
norepetidosTV = data[(data['show_title'].notnull()) & (~data['show_title'].astype(str).duplicated()) & (data.cumulative_weeks_in_top_10 > 10) & 
(data.country_iso2 == "AR") & (data.category == "TV")]
norepetidosTV = norepetidosTV.sort_values(by='cumulative_weeks_in_top_10', ascending=False)
norepetidosTV
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Calcular los pesos de las palabras basados en los valores numéricos
weights = dict(zip(norepetidosTV['show_title'], norepetidosTV['cumulative_weeks_in_top_10']))
# Crear la nube de palabras
wc = WordCloud(width = 1000, height = 300, background_color='black', colormap = "Reds").generate_from_frequencies(weights)

# Mostrar la nube de palabras
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```

```python
weights1 = dict(zip(norepetidos['show_title'], norepetidos['cumulative_weeks_in_top_10']))
# Crear la nube de palabras
wc1 = WordCloud(width = 1000, height = 300, background_color='black', colormap = "Reds").generate_from_frequencies(weights1)

# Mostrar la nube de palabras
plt.imshow(wc1, interpolation='bilinear')
plt.axis('off')
```
